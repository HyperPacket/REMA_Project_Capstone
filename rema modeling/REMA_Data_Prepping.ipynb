```python
#type of listing
type_map = {
    "apartment":0,
    "town house":1,
    "villas and palaces":2,
    "whole building":3,
    "farms and chalets":4
}
sale_df["type"] = sale_df["type"].map(type_map)
```

```python
from sklearn.preprocessing import LabelEncoder
city_enc = LabelEncoder()
sale_df["City"] = city_enc.fit_transform(sale_df["City"])
```

```python
from sklearn.preprocessing import LabelEncoder
import re

enc = LabelEncoder()
sale_df["Neighborhood"] = enc.fit_transform(sale_df["Neighborhood"])
```

```python
def convert_floor(text):
    text = str(text).lower()

    if "basement" in text:
      return -1
    if "ground" in text and "semi" in text:
      return 0.5
    if "ground" in text:
      return 0

    match = re.search(r"(\d+)", text)
    if match:
      return int(match.group(1))

    if "more than" in text:
      return 11

    return None  # if there isn't and for later

sale_df["floor"] = sale_df["floor"].apply(convert_floor)
```
